```python
def contract_start_year(x):
  if x == 'Free':
    return 'Free'
  elif "On Loan" in x:
    return "On Loan"
  else:
    match = re.search('(\d{4})\s*', x)
    if match:
      return match.group()
    else:
      return None
df['contract_start_year'] = df['Contract'].apply(contract_start_year)
df.head(2)
```

```python
def contract_end_year(x):
  if x == 'Free':
    return 'Free'
  elif "On Loan" in x:
    return "On Loan"
  else:
    match = re.search('\d{4}$', x)
    if match:
      return match.group()
    else:
      return None
df['contract_end_year'] = df['Contract'].apply(contract_end_year)
df.head(2)
```

```python
def height_to_cm(x):
  if "\'" in x:
    feet,inches = x.split("\'")
    ht = int(round(int(feet)*30.48 + int(inches.strip('"'))*2.54,0))
  else:
    ht = int(x.strip('cm'))
  return ht

df['Height'] = df['Height'].apply(height_to_cm)
```

```python
#since already the weight is converted to int with kgs, function will not run
def weight_to_kg(x):
  if "lbs" in x:
    weight = int(round(int(x.strip("lbs"))*0.453592 ,0))
  else:
    weight = int(x.strip('kg'))
  return weight

df['Weight'] = df['Weight'].apply(weight_to_kg)
```

```python
def value_to_euros(x):
  multipler=1
  if 'M' in x:
    multipler = 1000000
  elif 'K' in x:
    multipler = 1000
  value = x.strip('â‚¬').replace('M','').replace('K','')
  return int(float(value)*multipler)

df['Value'] = df['Value'].apply(value_to_euros)
#print(value_to_euros('â‚¬33.5M'))
```

```python
def wage_to_euros(x):
    multiplier = 1
    if 'K' in x:
      multiplier = 1000
    wage = x.strip('â‚¬').replace('K','')
    return int(wage)*multiplier

df['Wage'] = df['Wage'].apply(wage_to_euros)
```
